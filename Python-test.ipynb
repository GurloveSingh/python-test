```python
index = sbin.index
```

```python
np.asarray(index) # converting index into array form
```

```python
#create dummy time series Volume shock

Volume_arr= np.asarray(sbin["Volume"])


Volume_shock =[]
n = len(Volume_arr)
for i in range(n-1):
    if((Volume_arr[i+1])>1.1*(Volume_arr[i])):
        Volume_shock.append(0)
    else:
        Volume_shock.append(1)
Volume_shock.append(1)
new_df = pd.concat([pd.DataFrame(Volume_shock,index=index,columns=['Volume_shock']),sbin],axis=1)
new_df
```

```python
new_df['Volume_shock'].plot(kind='line',figsize=(20,10),color='b')
```

```python
#creating time series price shock
price_arr= np.asarray(sbin['VWAP'])

Price_shock =[]
n = len(price_arr)
for i in range(n-1):
    if((price_arr[i]-price_arr[i+1])>(0.02)):
        Price_shock.append(0)
    else:
        Price_shock.append(1)
        
Price_shock.append(1)
```

```python
len(Price_shock)
```

```python
new_df = pd.concat([pd.DataFrame(Price_shock,index=index,columns=['Price_shock']),new_df],axis=1)
```

```python
#Creating Price Shock Time Series
new_df['Price_shock'].plot(kind='line',figsize=(20,10),color='b')
```
